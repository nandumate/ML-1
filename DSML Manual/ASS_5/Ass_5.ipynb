```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
Data=pd.read_excel("/N_DSML.xlsx")
Data.head()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(Data['Age'], Data['RestBP'], c=Data['AHD'].map({'No': 'blue', 'Yes': 'red'}))
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='AHD', y='RestBP', data=Data, palette='viridis')
plt.title('Bar Plot')
plt.xlabel('Heart Disease (AHD)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
 #Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Chol', data=Data, palette='Set3')
plt.title('Box Plot')
plt.xlabel('Heart Disease (AHD)')
plt.ylabel('Cholesterol Level')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(Data['Age'], bins=10, color='yellow', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#change style using factorial,linewidth and edge color
plt.figure(figsize=(15,7))
sns.countplot(x="Age",data =Data,linewidth=3,facecolor=(0,0,0,0),
                 edgecolor=sns.color_palette("dark",10),palette="deep")
```

```python
# Pair Plot for Numerical Columns
numerical_columns = ['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak']
sns.pairplot(Data[numerical_columns])
plt.suptitle('Pair Plot for Numerical Columns', y=1.02)
plt.show()
```

```python
# Heatmap
correlation_matrix = Data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Fbs', y='Chol', data=Data, palette='dark:salmon_r')
plt.title('Swarm Plot for Fasting Blood Sugar (Fbs) and Cholesterol Level')
plt.xlabel('Fasting Blood Sugar (Fbs)')
plt.ylabel('Cholesterol Level')
plt.show()
```

```python
# Violin Plot for 'Age' and 'RestBP'
plt.figure(figsize=(12, 8))
sns.violinplot(x='AHD', y='Age', data=Data, palette='pastel')
plt.title('Violin Plot for Age and Heart Disease (AHD)')
plt.xlabel('Heart Disease (AHD)')
plt.ylabel('Age')
plt.show()
```

```python
# Bubble Plot for 'Age', 'RestBP', 'Chol' with 'MaxHR' as bubble size
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='RestBP', size='MaxHR', hue='Chol', sizes=(30, 300), data=Data, palette='viridis')
plt.title('Bubble Plot for Age, RestBP, Chol, and MaxHR')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (RestBP)')
plt.show()
```

```python
ahd_counts = Data['AHD'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(ahd_counts, labels=ahd_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Heart Disease (AHD)')
plt.show()
```
