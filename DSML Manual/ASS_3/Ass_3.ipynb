```python
#importing all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data
sales =pd.read_excel("/content/supermarket_sales.xlsx")
sales.head()
```

```python
sales.describe()
```

```python
sales.info()
```

```python
sales.shape
```

```python
#create a basic countplot to get the number of transaction for each product line
plt.figure(figsize=(15,8))
sns.countplot(x="Product line",data =sales)
```

```python
#make it horizontal bar plot
sns.countplot(y="Product line",data =sales)
```

```python
#to get count of to categories -Product line & Gender
#hue
plt.figure(figsize=(15,7))
sns.countplot(x="Product line",hue="Gender",data =sales)
```

```python
# using diffrent color palette
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data =sales,palette="deep")

```

```python
#change style using factorial,linewidth and edge color
plt.figure(figsize=(15,7))
sns.countplot(x="Product line",data =sales,linewidth=5,facecolor=(0,0,0,0),
                 edgecolor=sns.color_palette("dark",20),palette="deep")

```

```python
#plot heatmap using matplotlib
#imshow()=this function is used for creating heatmap
data =np.random.random((10,10))
plt.imshow(data,cmap="hot",interpolation="nearest")
plt.title("heatmap")
plt.colorbar()
plt.show()
```

```python
#implement a boxplot
#create a basic boxplot on total numeric variable
sns.boxplot(y="Total",data=sales,width=0.5)
```

```python
#create a boxplot on numeric variable by a categorical variable
#showmeans= to add MEAN marker in the box plot
sns.boxplot(x="Payment",y="Total",hue="Gender",data=sales,showmeans=True)
```
