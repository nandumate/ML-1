```python
import pandas as pd
N_DSML =pd.read_excel("/content/N_DSML.xlsx")
#individual functions
min_values = N_DSML.min(numeric_only=True)

max_values =  N_DSML.max(numeric_only=True)

# Mean
mean_values =  N_DSML.mean(numeric_only=True)



# Standard deviation
std_deviation =  N_DSML.std(numeric_only=True)

# Variance
variance =  N_DSML.var(numeric_only=True)
print("Minimum Values:\n", min_values)
print("\nMaximum Values:\n", max_values)
print("\nMean Values:\n", mean_values)
print("\nStandard Deviation:\n", std_deviation)
print("\nVariance:\n", variance)

```

```python
#b) Data Visualization-Create a histogram for each feature in the dataset to illustrate the
#feature distributions.
import matplotlib.pyplot as plt
N_DSML.hist(figsize=(15, 10), bins=40)
plt.suptitle("Histograms for Features", y=1.02)
plt.show()
```
